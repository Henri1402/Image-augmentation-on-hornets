```python
import os
import cv2
import imgaug.augmenters as iaa
```

```python
# Set the name of the folder containing the original images on your desktop
folder_name = 'hornets'
```

```python
# Set the path to the input folder containing original images
input_folder = os.path.expanduser(f'~/Desktop/bees&hornets/hornets')
```

```python
# Set the path to the output folder to save augmented images
output_folder = os.path.expanduser('~/Desktop/bees&hornets/augmented hornets')
```

```python
# Define the number of augmented images to generate per original image
num_augmented_images = 10
```

```python
# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)
```

```python
# Define augmentation transformations
augmenter = iaa.Sequential([
    iaa.Affine(rotate=(-45, 45)),  # Random rotation between -45 to 45 degrees
    iaa.Affine(scale={"x": (0.8, 1.2), "y": (0.8, 1.2)}),  # Random zoom in/out
    iaa.Affine(translate_percent={"x": (-0.2, 0.2), "y": (-0.2, 0.2)})  # Random translation
])
```

```python
# Loop through each image in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(('.jpg', '.png', '.jpeg')):  # Specify the image file extensions
        # Load the original image
        image = cv2.imread(os.path.join(input_folder, filename))

        # Generate augmented images
        for i in range(num_augmented_images):
            augmented_image = augmenter.augment_image(image)

            # Generate a unique filename for each augmented image
            augmented_filename = os.path.splitext(filename)[0] + f'_augmented_{i}.jpg'

            # Save the augmented image to the output folder
            cv2.imwrite(os.path.join(output_folder, augmented_filename), augmented_image)
```
